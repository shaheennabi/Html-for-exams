Q.1 Create an html page with red background with a message "warning" in large size blinking. Add scrolling text "read the message" below it.

```python
<html>
  <head>
    <style>
      .warning {
        background-color: red;
        font-size: 72px;
        text-align: center;
        animation: blink 1s linear infinite;
      }
      @keyframes blink {
        0% {
          opacity: 1;
        }
        50% {
          opacity: 0;
        }
        100% {
          opacity: 1;
        }
      }
      .scroll {
        font-size: 36px;
        text-align: center;
        overflow: hidden;
        white-space: nowrap;
        animation: scroll 10s linear infinite;
      }
      @keyframes scroll {
        0% {
          transform: translateX(0);
        }
        100% {
          transform: translateX(-100%);
        }
      }
    </style>
  </head>
  <body>
    <div class="warning">WARNING</div>
    <div class="scroll">read the message</div>
  </body>
</html>
```
